```python
import pandas as pd
import yfinance as yf
```

```python
ticker_list = ['xom', 'cvx', 'rydaf', 'ptr', 'tte', 'cop', 'eqnr', 'bp', 'enb', 'pbr']
```

```python
oil_companies_stock_prices = []
for t in ticker_list:
    ticker = yf.Ticker(t)
    if len(oil_companies_stock_prices) != 0:
        oil_companies_stock_prices.insert(len(oil_companies_stock_prices.columns), t, ticker.history(period='max')['Close'])
    else:
        oil_companies_stock_prices = pd.DataFrame(ticker.history(period='max')['Close'])
        oil_companies_stock_prices.columns = [t]
```

```python
oil_companies_stock_prices.to_csv('oil_companies_stock_prices.csv')
```
